```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
dat = np.loadtxt("Auger/lenBvsKEbinary2.txt")
```

```python
plt.figure(figsize=(20,20))
plt.plot(dat)
plt.xlim(0,8000)
plt.ylim(0,.8)
plt.title('Lens B vs binary voltage')
```

```python
plt.figure(figsize=(20,20))
plt.plot(np.diff(dat))
```

```python
bits = np.arange(65536)
```

```python
m, b = np.polyfit(bits, dat, 1)
B_noslope = dat - m*bits+b
print(m,b)
```

```python
plt.figure(figsize=(20,20))
plt.plot(B_noslope)
plt.ylim(-.12,0.01)
plt.xlim(0,8000)
```

```python
dat2 = np.loadtxt("Auger/outer_vsKEbinary.txt")
```

```python
plt.figure(figsize=(20,20))
plt.plot(dat2)
plt.xlim(0,8000)
plt.ylim(-.5,0)
plt.title('Outer hemisphere vs binary voltage')
```

```python
m2, b2 = np.polyfit(bits, dat2, 1)
Outer_noslope = dat2 - m2*bits+b2
print(m2,b2)
```

```python
plt.figure(figsize=(20,20))
plt.plot(Outer_noslope,label='outer hemisphere')
plt.plot(B_noslope,label='Lens B')
plt.ylim(-.12,0.05)
plt.legend()
plt.title('Slope-subtracted lens voltages vs binary code')
plt.xlim(0,8000)
```
