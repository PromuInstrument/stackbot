```python
import time
from pygrabber.dshow_graph import FilterGraph
import matplotlib.pyplot as plt
%matplotlib inline

#fig = plt.figure()
img_buffer = []

def img_callback(img):
    img_buffer.append(img)
    #print(img)
    #fig.imshow(img)

graph = FilterGraph()
print(graph.get_input_devices())
graph.add_input_device(0)

print("formats:")
for x in graph.get_formats():
    print( "\t", x)
graph.set_format(30)
#graph.add_default_render()
graph.add_sample_grabber(img_callback)
graph.add_null_render()
graph.prepare()
graph.run()
time.sleep(1)
graph.grab_frame()
```
