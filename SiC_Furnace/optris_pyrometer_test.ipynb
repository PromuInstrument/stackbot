```python
def read_temp():
    ser.write(b"\x01")
    x = ser.read(2)
    return 0.1*(x[0]*0xFF + x[1] - 1000)
    
```

```python
with serial.Serial("COM5", 115200, bytesize=8, parity='N', stopbits=1,timeout=0.1) as ser:
    print(read_temp())
    print(read_temp())
    print(read_temp())
```
